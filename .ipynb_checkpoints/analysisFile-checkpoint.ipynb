```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
```

```python
df=pd.read_csv("newSuperstore.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.set_index('Row ID')
```

```python
df.head()
```

```python
file_size = os.path.getsize('newSuperstore.csv')
file_size_mb = file_size / (1024 * 1024)
print(f"File size: {file_size_mb:.2f} MB")
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
new_dtypes = {
    'Country': 'category',
    'City': 'category',
    'State': 'category',
    'Region': 'category'
}
df=df.astype(new_dtypes)
```

```python
file_size = os.path.getsize('newSuperstore.csv')
file_size_mb = file_size / (1024 * 1024)
print(f"File size: {file_size_mb:.2f} MB")
```

```python
df.isnull().sum()
```

```python
df["Sales"]=df["Sales"].round(2)
df["Profit"]=df["Profit"].round(2)
```

# ORDER CATEGORY COUNT

```python
df["Category"].value_counts()
```

```python
plt.figure(figsize=(5,5))
count_plot=sns.countplot(x="Category",data=df,palette='viridis')
# Annotate each bar with the exact value
for bar in count_plot.containers:
    count_plot.bar_label(bar)
plt.title("Category count")
plt.savefig("Category count.jpg")
```

# SHIP MODE

```python
df["Ship Mode"].value_counts()
```

```python
x=df["Ship Mode"].value_counts().index
y=df["Ship Mode"].value_counts().values
print(x)
print(y)
```

```python
plt.figure(figsize=(4,4))
piePlot=plt.pie(y,labels=x,startangle=45,autopct="%0.2f%%",shadow = True)
plt.title("Shiping Mode")
plt.savefig("Shipping Mode.jpg")
```

```python
plt.figure(figsize=(7,5))
shipCountPlot=sns.countplot(x="Ship Mode",data=df,hue="Category")
for bar in shipCountPlot.containers:
    shipCountPlot.bar_label(bar)
plt.title("Category Wise Shipping Mode")
plt.savefig("Category Wise Shipping Mode.jpg")
```

# Customer Segment

```python
df["Segment"].value_counts()
```

```python
plt.figure(figsize=(5,5))
customerSegment=sns.countplot(x="Segment",data=df)
for bar in customerSegment.containers:
    customerSegment.bar_label(bar)
plt.title("Customer Segment")
```

```python
plt.figure(figsize=(5,4))
catCustomerSegment=sns.countplot(x="Segment",data=df,hue="Category")
for bar in catCustomerSegment.containers:
    catCustomerSegment.bar_label(bar)
plt.title("Category wise Customer Segment")
```
